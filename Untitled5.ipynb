```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
covid_df = pd.read_csv('covid_19_data.csv')
covid_df.head()
```

```python
covid_df = covid_df.drop(['SNo', 'Last Update'], axis=1)
covid_df.head()
```

```python
covid_df.info()
```

```python
covid_df_cleaned = covid_df.dropna()
covid_df_cleaned.info()
```

```python
covid_df_cleaned['ObservationDate'] = pd.to_datetime(covid_df_cleaned['ObservationDate'])
covid_df_cleaned.info()
```

```python
# Agrupar os dados por data e somar os valores globais
df_grouped = covid_df_cleaned.groupby("ObservationDate").sum()

# Criar o gráfico de linhas
plt.figure(figsize=(10,5))
plt.plot(df_grouped.index, df_grouped["Confirmed"], label="Casos Confirmados", color="blue")
plt.plot(df_grouped.index, df_grouped["Deaths"], label="Óbitos", color="red")
plt.plot(df_grouped.index, df_grouped["Recovered"], label="Recuperados", color="green")

# Personalizar o gráfico
plt.xlabel("Data")
plt.ylabel("Número de Casos")
plt.title("Evolução dos Casos Confirmados, Mortes e Recuperações")
plt.legend()
plt.xticks(rotation=45)
plt.grid()

# Exibir
plt.show()
```

```python
import matplotlib.pyplot as plt

# Remover a coluna de data antes da soma
df_country = covid_df_cleaned.groupby("Country/Region")[["Confirmed", "Deaths", "Recovered"]].sum()

# Ordenar pelos países com mais casos confirmados
df_top_countries = df_country.sort_values(by="Confirmed", ascending=False).head(10)

# Criar o gráfico de barras
plt.figure(figsize=(12,6))
plt.bar(df_top_countries.index, df_top_countries["Confirmed"], color="blue")

# Personalizar
plt.xlabel("País")
plt.ylabel("Casos Confirmados")
plt.title("Top 10 Países com Mais Casos Confirmados")
plt.xticks(rotation=45)
plt.grid(axis="y")

# Exibir
plt.show()
```

```python
import matplotlib.pyplot as plt

# Agrupar os dados por país e somar apenas as colunas numéricas
df_deaths = covid_df_cleaned.groupby("Country/Region")[["Deaths"]].sum()

# Selecionar os 5 países com mais mortes
df_top_deaths = df_deaths.sort_values(by="Deaths", ascending=False).head(5)

# Criar gráfico de pizza
plt.figure(figsize=(8,8))
plt.pie(df_top_deaths["Deaths"], labels=df_top_deaths.index, autopct="%1.1f%%", colors=["red", "orange", "gray", "blue", "green"])

# Personalizar
plt.title("Distribuição de Óbitos por País")

# Exibir
plt.show()
```
